```python
from jobspy import scrape_jobs
import pandas as pd
from IPython.display import display, HTML
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 50)
```

```python
# example 1 (no hyperlinks, India)
jobs = scrape_jobs(
    site_name=["linkedin"],
    location='India',
    search_term="Cyber Security",
    results_wanted=5,
)
display(jobs)
```

```python
# example 2 - remote India & hyperlinks
jobs = scrape_jobs(
    site_name=["linkedin", "indeed"],
    search_term="software engineer",
    country_indeed="India",
    hyperlinks=True,
    is_remote=True,
    results_wanted=5, 
)
display(jobs)
```

```python
# example 3 - with hyperlinks, india - linkedin with easy apply
jobs = scrape_jobs(
    site_name=["linkedin"],
    location='India',
    search_term="engineer",
    hyperlinks=True,
    results_wanted=5,
    easy_apply=True
)
display(jobs)
```
